```
import seaborn as sns
import matplotlib.pyplot as plt
import numpy
import pandas as pd
import os
from altair import Chart
from IPython.display import display
import altair as alt






class ZillowAnalysis():
    def __init__(self, filename):
      self.filename = filename
      #print("\n The file being analysed is " , self.filename)
      #data_rent = pd.read_csv(filename_rent , encoding='latin-1')
      #data_rent.head()
    def read_csv(self):
      try:
        self.df = pd.read_csv(self.filename)
      except IndexError:
        print("Error: Wrong file name")
        sys.exit(2)
      except UnicodeDecodeError:
        self.df = pd.read_csv(self.filename , encoding='latin-1')
      return self.df 
    def display_file(self):
      #print(self.df.columns)
      display(self.df.head())
      #print(self.df)
    def data_clean_by_year(self,year):
      self.year = str(year)
      #print("Dropping data create tidy data set, dropiing below year: " , self.year)
      self.df = self.df[self.df.columns.drop(list(filter(lambda x: x < self.year , list(self.df))))]
    def data_clean_by_column(self,*args):
      for i in args:
        self.columnName = str(i)
        #print("To create tidy data set, dropping column:" , self.columnName)
        self.df = self.df.drop([self.columnName], axis=1)
    def data_yearwise_mean(self,start,end,appendChar):
        self.start = start
        self.end = end
        self.appendChar = "col_"+str(appendChar)

        #print(self.start, self.end,self.appendChar)
        filter_text = ['RegionName']
        for x in range(self.start,self.end):
          self.appendChar = str(x)+appendChar
          ax = str(x)+"-01"
          ay = str(x)+"-12"
          #print(ax,ay,self.appendChar)
          col = self.df.loc[: , ax : ay]
          #print(type(self.appendChar))
          self.df[self.appendChar] = col.mean(axis = 1)
          filter_text.append(self.appendChar)
        #print(filter_text)
        self.df = self.df.filter(filter_text , axis = 1)
        return self.df

    def data_join_sale_rent(self,other):
      self.df = self
      self.other = other
      print(self.df,self.other,type(self.df),type(self.other))
      self.df = self.df.join(self.other, lsuffix='_caller', rsuffix='_other')
      return self.df

  

url_rent_metro = 'https://raw.githubusercontent.com/spurthikottan/fp_zillow_files/master/Metro_rent.csv'
url_sale_metro = 'https://raw.githubusercontent.com/spurthikottan/fp_zillow_files/master/Metro_sale.csv'

url_rent_state = 'https://raw.githubusercontent.com/spurthikottan/fp_zillow_files/master/state_rent.csv'
url_sale_state = 'https://raw.githubusercontent.com/spurthikottan/fp_zillow_files/master/state_sale.csv'

data_rent_metro = ZillowAnalysis(url_rent_metro)
data_sale_metro = ZillowAnalysis(url_sale_metro)

data_rent_state = ZillowAnalysis(url_rent_state)
data_sale_state = ZillowAnalysis(url_sale_state)



data_rent_state.read_csv()
#data_rent.display_file()
data_rent_state.data_clean_by_year(2011)
#data_rent.display_file()
data_rent_state.data_clean_by_column('SizeRank', 'RegionID')
#data_rent.display_file()
df_rent_state = data_rent_state.data_yearwise_mean(2012,2019,'rent')
#data_rent_state.display_file()

data_sale_state.read_csv()
#data_sale.display_file()
data_sale_state.data_clean_by_year(2011)
#data_sale.display_file()
data_sale_state.data_clean_by_column('SizeRank', 'RegionID')
#data_sale.display_file()
df_sale_state = data_sale_state.data_yearwise_mean(2012,2019,'sale')
#data_sale_state.display_file()

df_join_state = df_sale_state.join(df_rent_state, lsuffix='_caller', rsuffix='_other')
#display(df_join_state.head())
df_join_state = df_join_state.rename({'RegionName_caller' : 'RegionName'} , axis = 1)
df_join_state['2018sale'] = df_join_state['2018sale'].round(decimals=0)
df_join_state['2018rent'] = df_join_state['2018rent'].round(decimals=0)


source = df_join_state

rent_axis = '2018rent'
sale_axis = '2018sale'

alt.Chart(source, width = 1000 , height = 400).mark_point(color='green',size = 60).encode(
    
    x=alt.X(sale_axis, axis=alt.Axis(title='$$ Median Home Cost')),
    y=alt.Y(rent_axis, axis=alt.Axis(title='$$ Median Home Rent')),
    #size=alt.Size(rent_axis),
    tooltip=['RegionName', sale_axis , rent_axis ]
).properties(
    title='US median Home Rent vs Home Cost By State'
).interactive()
```

```
source = df_join_metro
rent_axis = '2018rent'
sale_axis = '2018sale'
region = 'state'


stripplot =  alt.Chart(source , width = 35 , height = 500).mark_circle(size=10).encode(
    x=alt.X(
        'jitter:Q',
        title=None,
        axis=alt.Axis(values=[0], ticks=True, grid=False, labels=False),
        
        scale=alt.Scale(),
    ),
    y=alt.Y('2018rent:Q', axis=alt.Axis(title='$$ Median Home Rent')),
    color=alt.Color('state:N', legend=None),
    size=alt.Size(sale_axis),
    column=alt.Column(
        'state:N',
        header=alt.Header(
            labelAngle=-90,
            titleOrient='bottom',
            labelOrient='bottom',
            labelAlign='right',
            labelPadding=3,
        ),
    ),
    tooltip=['RegionName', sale_axis , rent_axis ]
).transform_calculate(
    # Generate Gaussian jitter with a Box-Muller transform
    jitter='sqrt(-2*log(random()))*cos(2*PI*random())'
).configure_facet(
    spacing=0
).configure_view(
    stroke=None
).properties(
    title='US median Home Rent vs Home Cost by Metro categorized by State'
).interactive()

stripplot
```
